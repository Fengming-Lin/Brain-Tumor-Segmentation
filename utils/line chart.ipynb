```python
import matplotlib.pyplot as plt
import json
import numpy as np
```

```python
epoch = []
for i in range(70):
    epoch.append(i+1)
print(epoch)


with open('../weights/DUnet/0/dunet_tfcv5.json', 'r') as load_f:
    net1 = json.load(load_f)

with open('../weights/FCNN/0/fcnn_tfcv5.json', 'r') as load_f:
    net2 = json.load(load_f)

with open('../weights/VGG/0/VGG16_tfcv5.json', 'r') as load_f:
    net3 = json.load(load_f)
    
with open('../weights/PA+FP/0/PA+FP_tfcv5.json', 'r') as load_f:
    net4 = json.load(load_f)
    
with open('../weights/PA+FP+ED/0/dunet_tfcv5.json', 'r') as load_f:
    net5 = json.load(load_f)

with open('../weights/PA+EFP+ED/0/dunet_tfcv5.json', 'r') as load_f:
    net6 = json.load(load_f)
```

```python
plt.figure(figsize=(20, 15))


fontsize = 20
font1 = {'family': 'Times New Roman',
         'weight': 'normal',
         'size': fontsize}



plt.subplot(3,3,1)
plt.plot(epoch, net4['val_dice_comp'],label='PA+FP(Ours)')
plt.plot(epoch, net1['val_dice_comp'],label='No PA+FP')
plt.legend(prop=font1)
plt.xlabel('Epoch', fontsize=fontsize)
plt.ylabel('Dice', fontsize=fontsize)
plt.title('Whole Tumor', fontsize=fontsize)

plt.subplot(3,3,2)
plt.plot(epoch, net4['val_dice_core'],label='PA+FP(Ours)')
plt.plot(epoch, net1['val_dice_core'],label='No PA+FP')
plt.legend(prop=font1)
plt.xlabel('Epoch', fontsize=fontsize)
plt.ylabel('Dice', fontsize=fontsize)
plt.title('Tumor Core', fontsize=fontsize)

plt.subplot(3,3,3)
plt.plot(epoch, net4['val_dice_en'],label='PA+FP(Ours)')
plt.plot(epoch, net1['val_dice_en'],label='No PA+FP')
plt.legend(prop=font1)
plt.xlabel('Epoch', fontsize=fontsize)
plt.ylabel('Dice', fontsize=fontsize)
plt.title('Enhancing Tumor', fontsize=fontsize)



plt.subplot(3,3,4)
plt.plot(epoch, net5['val_dice_comp'],label='PA+FP+ED(Ours)')
plt.plot(epoch, net1['val_dice_comp'],label='No PA+FP+ED')
plt.legend(prop=font1)
plt.xlabel('Epoch', fontsize=fontsize)
plt.ylabel('Dice', fontsize=fontsize)
plt.title('Whole Tumor', fontsize=fontsize)

plt.subplot(3,3,5)
plt.plot(epoch, net5['val_dice_core'],label='PA+FP+ED(Ours)')
plt.plot(epoch, net1['val_dice_core'],label='No PA+FP+ED')
plt.legend(prop=font1)
plt.xlabel('Epoch', fontsize=fontsize)
plt.ylabel('Dice', fontsize=fontsize)
plt.title('Tumor Core', fontsize=fontsize)

plt.subplot(3,3,6)
plt.plot(epoch, net5['val_dice_en'],label='PA+FP+ED(Ours)')
plt.plot(epoch, net1['val_dice_en'],label='No PA+FP+ED')
plt.legend(prop=font1)
plt.xlabel('Epoch', fontsize=fontsize)
plt.ylabel('Dice', fontsize=fontsize)
plt.title('Enhancing Tumor', fontsize=fontsize)



plt.subplot(3,3,7)
plt.plot(epoch, net6['val_dice_comp'],label='PA+FP+ED(Ours)')
plt.plot(epoch, net1['val_dice_comp'],label='No PA+EFP+ED')
plt.legend(prop=font1)
plt.xlabel('Epoch', fontsize=fontsize)
plt.ylabel('Dice', fontsize=fontsize)
plt.title('Whole Tumor', fontsize=fontsize)

plt.subplot(3,3,8)
plt.plot(epoch, net6['val_dice_core'],label='PA+FP+ED(Ours)')
plt.plot(epoch, net1['val_dice_core'],label='No PA+EFP+ED')
plt.legend(prop=font1)
plt.xlabel('Epoch', fontsize=fontsize)
plt.ylabel('Dice', fontsize=fontsize)
plt.title('Tumor Core', fontsize=fontsize)

plt.subplot(3,3,9)
plt.plot(epoch, net6['val_dice_en'],label='PA+FP+ED(Ours)')
plt.plot(epoch, net1['val_dice_en'],label='No PA+EFP+ED')
plt.legend(prop=font1)
plt.xlabel('Epoch', fontsize=fontsize)
plt.ylabel('Dice', fontsize=fontsize)
plt.title('Enhancing Tumor', fontsize=fontsize)


plt.tight_layout(1, 2)

# plt.savefig('../results/figures/2018chart.eps')
plt.show()
```
